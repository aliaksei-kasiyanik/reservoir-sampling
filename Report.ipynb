# C1207 - Reservoir Sampling

## Задача

В рамках данной лабораторной работы была выполнена серия экспериментов с целью изучения потоковых алгоритмов получения случайной выборки (sampling). 

Были реализованы алгоритмы для получения равновероятностной и взвевешенной выборок.

В случае равновероятностной выборки ожидается, что любой элемент потока может попасть в выборки с одинаковой вероятностью, т.е. если $n$ - количество элементов в потоке, то вероятность попадания элемента - $\frac{1}{n}$.

В случае взвешенной выборки были рассмотрены два случая: 
1. $w_{i} = i$ - линейная зависимость
2. $w_{i} = i^{2}$ - квадратичная зависимость

Результаты проведенной серии экспериментов (10000 экспериментов) для различного количества элементов в потоке и различного размера выборки совпадают с ожидаемыми и тем самым подтверждают корректную работу алгоритмов.

## Результаты

Код генерации графиков: 

```python
from os import listdir, path
import re
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plot_results(path_to_results_dir):
    for filename in listdir(path_to_results_dir):
        filename_parts = re.split("[-.]", filename)
        method = filename_parts[0]
        n = int(filename_parts[1])
        m = int(filename_parts[2])
        experiments = int(filename_parts[3])
        with open(path.join(path_to_results_dir, filename), 'r') as f:
            content = f.readlines()
            results = list(map(int, content))
            items = range(n)
            plt.figure(figsize=(15, 5), dpi=600)
            plt.title('method = {}, n = {}, m = {}, experiments = {}'.format(method, n, m, experiments),fontsize=20)
            plt.ylabel("Occurrence in sample")
            plt.xlabel("Stream item")
            plt.grid(True)
            plt.bar(items, results, 1/1)
            plt.show()
```

Вывод результатов:

```python
plot_results("src/main/resources/results")
```
